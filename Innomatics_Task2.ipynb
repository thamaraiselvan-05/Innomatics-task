### 1) Given a valid (IPv4) IP address, return a defanged version of that IP address.

<p>A defanged IP address replaces every period "." with "[.]".</p>
<p>Input: address = "1.1.1.1"</p>
<p>Output: "1[.]1[.]1[.]1"</p>

```python
def defang_ip(address):
    return address.replace(".", "[.]")

address="1.1.1.1"
defang_ip(address)
```

### 2) Given an array nums of integers, return how many of them contain an even number of digits.
<p>Input: nums = [12,345,7896]</p>
<p>Output: 2 . Only 12 and 7896 contain an even number of digits.</p>
<p>Explanation: </p>
<p>12 contains 2 digits (even number of digits). </p>
<p>345 contains 3 digits (odd number of digits). </p>
<p>7896 contains 4 digits (even number of digits). </p>

```python
def find_even_digit_numbers(nums):
    count = 0
    for num in nums:
        if len(str(abs(num))) % 2 == 0:
            count += 1
    return count

nums = [12, 345, 2, 6, 7896]
print(find_even_digit_numbers(nums))
```

### 3) Given an array of integers nums, return the number of good pairs.

<p>A pair (i, j) is called good if nums[i] == nums[j] and i < j. </p>

Example 1:

Input: nums = [1,2,3,1,1,3]
<p>Output: 4</p>
<p>Explanation: There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.</p>

```python
nums = [1, 2, 3, 1, 1, 3]

count = 0

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] == nums[j]:
            count += 1

print(count)
```

### 4) Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].

<p>Return the answer in an array.</p>

 

<p>Example 1:</p>

<p>Input: nums = [8,1,2,2,3]</p>
<p></p>Output: [4,0,1,1,3]</p>
Explanation: 
<p>For nums[0]=8 there exist four smaller numbers than it (1, 2, 2 and 3). </p>
<p>For nums[1]=1 does not exist any smaller number than it.</p>
<p>For nums[2]=2 there exist one smaller number than it (1). </p>
<p>For nums[3]=2 there exist one smaller number than it (1). </p>
<p>For nums[4]=3 there exist three smaller numbers than it (1, 2 and 2).</p>

```python
nums = [8, 1, 2, 2, 3]
result = []

for i in nums:
    count = 0
    for j in nums:
        if j < i:
            count += 1
    result.append(count)
print(result)
```

### 5) Subtract the Product and Sum of Digits of an Integer
Example 1:
<p>Input: n = 234</p>
<p>Output: 15 </p>
<p>Explanation: </p>
<p>Product of digits = 2 * 3 * 4 = 24 </p>
<p>Sum of digits = 2 + 3 + 4 = 9 </p>
<p>Result = 24 - 9 = 15</p>

```python
n = 234
product = 1   
sum_digits = 0  
for digit in str(n):     
    digit = int(digit)  
    product = product * digit
    sum_digits = sum_digits + digit
answer = product - sum_digits
print(answer)
```

### 6) XOR Operation in an Array

<p>Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.</p>

<p>Return the bitwise XOR of all elements of nums.</p>

<p>Input: n = 5, start = 0</p>
<p>Output: 8</p>
<p>Explanation: Array nums is equal to [0, 2, 4, 6, 8] where (0 ^ 2 ^ 4 ^ 6 ^ 8) = 8.
Where "^" corresponds to bitwise XOR operator.</p>

```python
n = 5
start = 0
result = 0

for i in range(n):
    result = result ^ (start + 2*i)
print(result)
```
